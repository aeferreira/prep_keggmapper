# Preparação de dados para o KeGGMapper

Este _notebook_ utiliza os ficheiros gerados pelo MassTrix (para cada estirpe) e cria um ficheiro chamado demo.txt adequado para submeter à ferramenta KeGG Mapper: contem uma coluna dos identificadores KeGG dos compostos encontrados em ambas as estirpes e uma coluna com o seguinte código de cores:

- yellow - Compostos presentes em ambas as estirpes

- red - Compostos presentes na estirpe mutada (deltaX)

- green - Compostos presentes na estirpe wild type (BY0)


### Instruções:

0. Colocar os ficheiros .annotated de ambos as estirpes na mesma pasta que o notebook demo.ipynb

1. Mudar os nomes dos ficheiros .annotaded do MassTriX para

    - `masses_DX.annotated` para a estirpe delta
    - `masses_BY.annotated` para a estirpe selvagem (BY)


2. Correr a célula seguinte deste notebook.

3. É gerado um ficheiro **demo.txt**. Deve-se fazer copy-paste do conteúdo deste ficheiro para o KeGG Mapper, excluíndo a primeira linha

```python
with open("masses_BY.annotated") as a:
    allBY = a.read()

with open("masses_DX.annotated") as a:
    allDX = a.read()

DX = []
BY = []
for compound in allDX.splitlines():
    parts = compound.split("\t")
    ID = parts[6]
    if ID.startswith("C"):
        DX.append(ID)

for compound in allBY.splitlines():
    parts = compound.split("\t")
    ID = parts[6]
    if ID.startswith("C"):
        BY.append(ID)

allcompounds = DX + BY

allcompounds = list( dict.fromkeys(allcompounds))

text = "KEGG_ID\tcolor"

for compound in allcompounds:
    if compound in DX:
        vDX = "Sim"
    else:
        vDX = "Não"
    if compound in BY:
        vBY = "Sim"
    else:
        vBY = "Não"
    if vDX == "Sim" and vBY == "Sim":
        color = "yellow"
    if vDX == "Sim" and vBY == "Não":
        color = "red"
    if vDX == "Não" and vBY == "Sim":
        color = "green"
        
    text += f"\n{compound}\t{color}"

with open("demo.txt", "w+") as a:
    a.write(text)
```
